# 1. Preliminaries

```python
import pickle                           # Load and save pickle files.
import datetime                         # Handle date/time objects.
import re                               # Regular expressions.
import matplotlib as mpl
import matplotlib.pyplot as plt         # Plot things.
import numpy as np                      # Basic math and array operations.
from scipy.interpolate import interp1d  # Interpolation.
```

The following section of preliminaries imports the forecast data for hurricane Ian, extracts and interpolates the true path of the hurricane, and defines a function to compute distance between two points.

```python
# Import the forecast data I provided.
with open('forecasts_2022_ian.pkl','rb') as f:
    forecasts = pickle.load(f)

# Extract true path.
true_times      = [ f[0][0] for f in forecasts ]
true_latitudes  = [ f[0][1] for f in forecasts ]
true_longitudes = [ f[0][2] for f in forecasts ]

# Convert datetimes to seconds.
true_times_sec  = [ (t-true_times[0]).total_seconds() for t in true_times ]

# Interpolate true path.
longitude = interp1d(true_times_sec,true_longitudes)
latitude  = interp1d(true_times_sec,true_latitudes)

# Function that computes the distance (in miles) between two points on the globe given their GPS coordinates.
# lo1=longitude of first point, la1=latitude of first point
# lo2=longitude of second point, la2=latitude of second point
import pyproj
def geoDistance(la1,lo1,la2,lo2):
    return pyproj.Geod(ellps='clrk66').inv(lo1,la1,lo2,la2)[2]/1609
```

# 2. Average forecast error

In this part we want to compute the average distance between the hurricane's predicted location and its actual location as a function of the amount of time between the predicting and the prediction. Intuitively the distance should grow with the time difference: the further into the future the prediction, the less accurate.

To measure this, we need to look at every forecast location and compile 1) how far into the future the forecast was made (the "time difference" for short), 2) where the hurricane was predicted to be, 3) where the hurricane ended up being at the predicted time, and 4) the distance between 2) and 3), i.e., how far off the prediction was. As we do that, we need to build of list of 1) and 4). Once the list is made, we group the (time difference,distance) pairs by time and compute the average distance in each group. Finally we plot this average against the time difference.

## 2.1 Visual average

Next, I compute and plot the distance between the predicted and true location of the hurricane against the amount of time into the future the prediction was made. The predicted points whose date/time fall after the last point of the true path must be dismissed since the distance cannot be calculated without knowing the true location of the hurricane at that time, and interpolation does not work beyond the range of true points.  

```python
#list of colors for plotting forecasts
colors = mpl.cm.jet(np.linspace(0,1,len(forecasts)))
colors = [mpl.colors.rgb2hex(c) for c in colors]
```

```python
#i stores the index of current forecast, and f stores the list of predictions of current forecast
for i, f in enumerate(forecasts):
    distances = [] #empty lists for distances and days into the future
    days = []
    #t is a datetime object, la1 is the latitude value, lo2 is the longitude value of the current prediction of forecast f
    for t, la1, lo1 in f:
        dt = (t - forecasts[0][0][0]).total_seconds() #calculate time difference (seconds) between current forecast and the first forecast
        
        #Prevent interpolation of points that are beyond the range of the true hurricane path 
        if dt > true_times_sec[-1]: 
            continue
        
        #Calculate distance difference between prediction and location
        distances.append(geoDistance(la1, lo1, latitude(dt), longitude(dt)))
    
        #Calculate the time difference between current prediction and first prediction of f
        days.append((t - f[0][0]).total_seconds() / 86400)

    #Plot days against distances. Each forecast is assigned a specific color
    plt.plot(days, distances, marker='o', ms=3.5, color=colors[i], linestyle='-', lw=.5)
    #Titles and axis names
    plt.xlabel('Time Difference (days)')
    plt.ylabel('Distance (miles)')
    plt.title('Figure 1. Visual Average: Time Difference vs. Distance')
```

As expected, we can see from _Figure 1_ that the accuracy of the predictions declines the further into the future they are done. When attempting to predict the path of the hurricane 5 days into the future, the forecasts had an error of up to 300 miles away from its true path.

## 2.2 Actual average

<div class="alert alert-block alert-danger">
<b>Task 2b:</b>
</div>

Here, I organize the forecast data into a list of pairs of seconds into the future and location difference in miles. The predicted locations whose data/time is beyond the last true point of the hurricane are dismissed again for the aforementioned reasons.

```python
pairs = [] #empty list of pairs (seconds into the future, distance difference between prediction and true location)

for f in forecasts: #f stores the list of predictions of current forecast
    for t, la1, lo1 in f: #t is a datetime object, la1 is the latitude value, lo2 is the longitude value of the current prediction of forecast f
        dt = (t - forecasts[0][0][0]).total_seconds() #dt is the time difference in seconds
        if dt > true_times_sec[-1]: #prevents interpolation of points outside of range
            continue
        pairs.append((((t - f[0][0]).total_seconds()), (geoDistance(la1, lo1, latitude(dt), longitude(dt))))) #retrieve pair and store
```

<div class="alert alert-block alert-danger">
<b>Task 2c:</b>  
</div>

Now, I compute and plot the **average** location difference as a function of time difference in days.

```python
#unique_sf stores sorted, single instances of pairs[0]
unique_sf = np.sort(np.unique([sf[0] for sf in pairs]))

#the following nested for loops compare each unique time in unique_sf to every time in the list of pairs and if they are 
#equal it retrieves and stores the corresponding distance. Then, it calculates the average of distances whose times are the 
#same and stores it in the list avg

avg = [] #empty list that will store the average of the distance differences
for time in unique_sf:
    temp = [] #temporary list to store distance differences at a specific time difference
    for pair in pairs:
        if pair[0] == time:
            temp.append(pair[1]) #retrieve distances and store them
    avg.append(np.mean(temp)) #calculate average of the distances and store it
    
days = [s/86400 for s in unique_sf] #convert the unique seconds to days

#plot average distance against time difference
plt.plot(days, avg,  marker='o', ms=3.5, color='blue', linestyle='-', lw=.5)
plt.xlabel('Time Difference (days)')
plt.ylabel('Average Distance (miles)')
plt.title('Figure 2. Actual Average: Time Difference vs. Average Distance')
```

As illustrated by _Figure 2_, the forecasting error increases as the predictions are done further into the future.
